```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(5*x)
```

```python
plt.axes().set_facecolor("lightgreen")
plt.plot(x,y)
plt.ylim(-1.1,1.1)
plt.xlabel("x")
plt.ylabel("sin (5*x)")
plt.savefig("_trial.png", facecolor='lightblue')
plt.show()
```
