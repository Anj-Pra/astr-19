Create a Jupyter notebook, import matplotlib. Write cells that create an array x ranging from [0,1] in 100 steps and that defines functions that return sin(x) and cos(x).  In a new cell use to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos
```

```python
def dosin(x): # function that return sin(x)
    return(np.sin(x))
```

```python
def docos(x): # function that return cos(x)
    return(np.cos(x))
```

```python
x = np.arange(0,1.01,0.01) # array x ranging from [0,1] in 100 steps
sin_x = dosin(x)
cos_x = docos(x)
```

```python
f, axarr = plt.subplots(1,2) # multipanel plot (1 row, 2 columns)

axarr[0].plot(x,sin_x) # sin(x) vs. x
axarr[0].set_xlabel("x")
axarr[0].set_ylabel("sin(x)")
axarr[0].set_title(r"$\sin(x)$")

axarr[1].plot(x,cos_x) # cos(x) vs. x
axarr[1].set_xlabel("x")
axarr[1].set_ylabel("cos(x)")
axarr[1].set_title(r"$\cos(x)$")

f.subplots_adjust(wspace = 0.4)
f.savefig("figure2.pdf")
```
